```python
from ultralytics import YOLO
```

```python
model = YOLO('/content/drive/MyDrive/Human_Faces_Object_Detection/yolov8m.pt')
```

```python
model.train(data="/content/data.yaml", epochs=80, imgsz=1024, batch=16, name='face_yolo_exp')
```

```python
# Load your best model
model = YOLO("/content/drive/MyDrive/Human_Faces_Object_Detection/Faces_YOLO_Model/detect/face_yolo_exp/weights/best.pt")

# Evaluate on validation set
metrics = model.val(data="/content/data.yaml")   # uses data.yaml automatically
print(metrics)
```

```python
print("Precision:", metrics.box.p)
print("Recall:", metrics.box.r)
print("F1 Score:", metrics.box.f1)
print("mAP@0.5:", metrics.box.map50)
print("mAP@0.5:0.95:", metrics.box.map)
```

```python
df_metrics = pd.DataFrame({
    "Metric": ["Precision", "Recall", "F1", "mAP@0.5", "mAP@0.5:0.95"],
    "Value": [
        metrics.box.p[0],
        metrics.box.r[0],
        metrics.box.f1[0],
        metrics.box.map50,
        metrics.box.map
    ]
})
```
